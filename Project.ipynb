```python
# câu 1
def password(a): 
        a = list(a)
        b = ["!", "@", "#", "$"]  
        c = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
        C = [x.upper() for x in c]  
        d = []  
        
        if len(a) <8 or len(a)>15:
            print('Độ dài không hợp lệ')
            return False 
        
        number=upper=special=lower=False
        for i in range(0, len(a)):
            if a[i] in b:
                special=True
            elif a[i] in c:
                lower=True
            elif a[i] in C:
                upper=True
            elif a[i].isdigit():
                number=True
            else: 
                print('Không hợp lệ', a[i])
                return False
            
        if not special:
                print('Ko có ký tự đặc biệt')
                return False
        if not lower:
                print('ko có ký tự thường')
                return False
        if not upper: 
                print('ko có ký tự in hoa')
                return False
        if not number:
                print('Ko có số')
                return False 
        print(f'Mật khẩu {a} hợp lệ')
        return True
```
